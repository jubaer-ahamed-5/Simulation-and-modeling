```
#Tausworthe Genration
r=3
q=5
n=1000
l=4
b_i=[1,1,1,1,1]
for i in range(5,n):
  b = (b_i[i-r] + b_i[i-q])%2
  b_i.append(b)

  ##equation of new bit
  ##bit append to b_i list
print(b_i)  
```

```
u=[]
for i in range (0,len(b_i),l):
  new_u = int("".join(str(i) for i in b_i[i:i+l]),2)
  u.append(new_u)
print(u)
```

```
c_length = -1
for i in range(1,len(u)-1):
  if u[0]==u[i]:
    for j in range(1,i):
      if u[j]!=u[i + 1]:
        break
      c_length =i
      i=i+1
      if c_length!= -1:
        break
if c_length == -1:
  print("No cycle for " + str(n) + " random numbers")
else:
  print("Cycle found  for " + str(n) + " random numbers")
  print("the cycle length is=" + str(c_length))
```
